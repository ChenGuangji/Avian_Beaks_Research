```python
import pickle
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr 
from itertools import combinations
```

```python
filename = 'traits.pkl'
with open(filename, 'rb') as file:
    df = pickle.load(file)
```

```python
df
```

```python
df['mass_cr'] = df['Mass_gram']**(1/3)
df = df[df['mass_cr'] != 0] 
df.dropna(subset=['mass_cr'], inplace=True) 
df['BL'] = df['Beak_Length_Culmen_mm'] / df['mass_cr']
df['BW'] = df['Beak_Width_mm'] / df['mass_cr']
df['BD'] = df['Beak_Depth_mm'] / df['mass_cr']
```

```python
def calculate_corr_and_pvalue_matrices(df_subset, columns):
    """
    Calculates correlation coefficients and p-value matrices for given columns.
    """
    corr_matrix = pd.DataFrame(index=columns, columns=columns, dtype=float)
    pval_matrix = pd.DataFrame(index=columns, columns=columns, dtype=float)
    for col1 in columns:
        for col2 in columns:
            if col1 == col2:
                corr_matrix.loc[col1, col2] = 1.0
                pval_matrix.loc[col1, col2] = 0.0 # p-value for self-correlation
            elif pd.isna(corr_matrix.loc[col1, col2]): # Only calculate if not already filled (for symmetry)
                # Dropna for pairs to ensure pearsonr works correctly
                valid_data = df_subset[[col1, col2]].dropna()
                if len(valid_data) >= 2: # pearsonr requires at least 2 samples
                    r, p = pearsonr(valid_data[col1], valid_data[col2])
                else:
                    r, p = np.nan, np.nan # Not enough data
                
                corr_matrix.loc[col1, col2] = r
                corr_matrix.loc[col2, col1] = r # Symmetric
                pval_matrix.loc[col1, col2] = p
                pval_matrix.loc[col2, col1] = p # Symmetric
            
    return corr_matrix, pval_matrix

def get_significance_stars(p_value):
    """Returns significance stars based on p-value."""
    if pd.isna(p_value):
        return ""
    if p_value < 0.001:
        return "***"
    elif p_value < 0.01:
        return "**"
    elif p_value < 0.05:
        return "*"
    return "" # You can use "ns" for non-significant if preferred
```

```python
COLUMN_1 = 'BL'
COLUMN_2 = 'BW'
COLUMN_3 = 'BD'
GROUP_COLUMN = 'Clade'
data_columns = [COLUMN_1, COLUMN_2, COLUMN_3]

overall_corr_matrix, overall_pval_matrix = calculate_corr_and_pvalue_matrices(df, data_columns)

print("Overall Correlation Matrix:")
print(overall_corr_matrix)
print("\nOverall P-value Matrix:")
print(overall_pval_matrix)
print("\n" + "="*50 + "\n")

correlation_summary_list = []
group_names = df[GROUP_COLUMN].unique()
data_columns_for_corr = [COLUMN_1, COLUMN_2, COLUMN_3] # å’Œ data_columns ä¸€æ ·

for group_name in group_names:
    group_df = df[df[GROUP_COLUMN] == group_name]
    if len(group_df) < 2: # Need at least 2 samples for correlation
        print(f"Warning: Group '{group_name}' has less than 2 samples. Skipping its correlation summary.")
        continue

    g_corr_matrix, g_pval_matrix = calculate_corr_and_pvalue_matrices(group_df, data_columns_for_corr)

    summary = {'Group': group_name, 'N_samples': len(group_df)} # Storing sample size might be useful
    for var1_idx, var1_name in enumerate(data_columns_for_corr):
        for var2_idx, var2_name in enumerate(data_columns_for_corr):
            if var1_idx < var2_idx:
                corr_pair_name = f"corr({var1_name},{var2_name})"
                pval_pair_name = f"pval({var1_name},{var2_name})"
                
                summary[corr_pair_name] = g_corr_matrix.loc[var1_name, var2_name]
                summary[pval_pair_name] = g_pval_matrix.loc[var1_name, var2_name]
    correlation_summary_list.append(summary)

if not correlation_summary_list:
    print("Error: Failed to generate correlation summary data. Cannot proceed with plotting.")

correlation_summary_df = pd.DataFrame(correlation_summary_list)
if not correlation_summary_df.empty:
    correlation_summary_df.set_index('Group', inplace=True)
```

```python
print("Starting plotting...\n")

if 'correlation_summary_df' not in locals() or correlation_summary_df.empty:
    print("No valid correlation summary data available for plotting. Skipping plots.")
else:
    if overall_corr_matrix is not None and not overall_corr_matrix.empty:
        overall_annot_texts = pd.DataFrame(index=data_columns, columns=data_columns, dtype=str)
        for r_name in data_columns: # r_name is row name
            for c_name in data_columns: # c_name is col name
                r_val = overall_corr_matrix.loc[r_name, c_name]
                p_val = overall_pval_matrix.loc[r_name, c_name]
                stars = get_significance_stars(p_val) if r_name != c_name else "" # No stars on diagonal
                
                if pd.notna(r_val):
                    overall_annot_texts.loc[r_name, c_name] = f"{r_val:.2f}{stars}"
                else:
                    overall_annot_texts.loc[r_name, c_name] = "NA"


        plt.figure(figsize=(7, 6)) # Adjusted size for potentially longer annotations
        sns.heatmap(overall_corr_matrix, annot=overall_annot_texts, fmt="s", cmap="coolwarm", 
                    vmin=-1, vmax=1, center=0,
                    linewidths=.5, cbar_kws={"shrink": .8})
        plt.title('Overall Correlation Matrix (with Significance)', fontsize=14)
        plt.xticks(rotation=45, ha="right", fontsize=10)
        plt.yticks(rotation=0, fontsize=10)
        plt.tight_layout()
        overall_corr_plot_filename = "heatmap_overall_correlations_with_pvals.pdf" # New filename
        try:
            plt.savefig(overall_corr_plot_filename)
            print(f"Plot saved as: {overall_corr_plot_filename}")
        except Exception as e:
            print(f"Error saving overall correlation heatmap: {e}")
        plt.show()
    else:
        print("Overall correlation data is missing, cannot draw heatmap.")

    r_value_cols_plot2 = [col for col in correlation_summary_df.columns if col.startswith('corr(')]

    if not correlation_summary_df.empty and r_value_cols_plot2:
        heatmap_r_values_transposed = correlation_summary_df[r_value_cols_plot2].T

        custom_x_axis_order = ['Struthioniformes','Casuariiformes','Apterygiformes','Rheiformes','Tinamiformes',
                               'Galliformes','Anseriformes',
                               'Phoenicopteriformes','Podicipediformes',
                               'Mesitornithiformes','Pterocliformes','Columbiformes',
                               'Musophagiformes','Otidiformes','Cuculiformes',
                               #'Opisthocomiformes',
                               'Gruiformes','Charadriiformes','Strisores',
                               'Phaethontiformes','Eurypygiformes','Gaviiformes','Sphenisciformes','Procellariiformes','Pelecaniformes',
                               'Strigiformes','Accipitriformes','Coliiformes',#'Leptosomatiformes',
                               'Trogoniformes','Bucerotiformes','Coraciiformes','Piciformes',
                               'Cariamiformes','Falconiformes','Psittaciformes','Acanthisitti','Tyranni','Passeri'] # 
        
        available_groups_plot2 = heatmap_r_values_transposed.columns.tolist() # Groups are now columns
        ordered_groups_plot2 = [group for group in custom_x_axis_order if group in available_groups_plot2]
        for group in available_groups_plot2:
            if group not in ordered_groups_plot2:
                ordered_groups_plot2.append(group)
        
        if ordered_groups_plot2: # Ensure there's something to plot
            heatmap_r_values_ordered = heatmap_r_values_transposed.reindex(columns=ordered_groups_plot2)
        else:
            heatmap_r_values_ordered = heatmap_r_values_transposed # Fallback to original order if ordering fails
            print("Warning: Custom group ordering for Plot 2 might not have applied as expected.")

        annot_texts_plot2 = pd.DataFrame(index=heatmap_r_values_ordered.index, 
                                         columns=heatmap_r_values_ordered.columns, dtype=str)
        
        for corr_pair_name_idx in heatmap_r_values_ordered.index: 
            pval_col_name_original = corr_pair_name_idx.replace("corr(", "pval(")

            for group_name_col in heatmap_r_values_ordered.columns:
                r_val = heatmap_r_values_ordered.loc[corr_pair_name_idx, group_name_col]

                if pval_col_name_original in correlation_summary_df.columns:
                    p_val = correlation_summary_df.loc[group_name_col, pval_col_name_original]
                else:
                    p_val = np.nan # P-value column not found
                    print(f"Warning: P-value column '{pval_col_name_original}' not found for group '{group_name_col}'.")

                stars = get_significance_stars(p_val)
                
                if pd.notna(r_val):
                    annot_texts_plot2.loc[corr_pair_name_idx, group_name_col] = f"{r_val:.2f}{stars}"
                else:
                    annot_texts_plot2.loc[corr_pair_name_idx, group_name_col] = "NA"

        plt.figure(figsize=(14,5)) 
        sns.heatmap(heatmap_r_values_ordered, annot=annot_texts_plot2, fmt="s", cmap="coolwarm", 
                    vmin=-1, vmax=1, center=0, annot_kws={"fontsize":5},
                    linewidths=.5, cbar_kws={"shrink": .8, "label": "Pearson Correlation Coefficient (r)"})
        plt.title('Within-Group Trait Correlations (with Significance)', fontsize=14)
        plt.xlabel('Clade (Group)', fontsize=12) # Changed from Species to Clade
        plt.ylabel('Trait Correlation Pair', fontsize=12)
        plt.xticks(rotation=45, ha="right", fontsize=10)
        plt.yticks(rotation=0, fontsize=10)
        plt.tight_layout()
        within_group_corr_plot_filename = "heatmap_within_group_correlations_with_pvals.pdf" # New filename
        try:
            plt.savefig(within_group_corr_plot_filename)
            print(f"Plot saved as: {within_group_corr_plot_filename}")
        except Exception as e:
            print(f"Error saving within-group correlation heatmap: {e}")
        plt.show()
    else:
        print("Data for within-group correlation heatmap is empty or improperly formatted.")
```
