```r
library(ape)
library(treeio)
library(ggtree)
library(dplyr)
library(ggplot2)
library(tibble)
library(ggtreeExtra)
library(cowplot)
```

```r
tree <- readRDS(file = "./tree.rds")
traits <- readRDS(file = "./traits.rds")

common_sp <- intersect(tree$tip.label, traits$Species)


tree_pruned <- drop.tip(tree, setdiff(tree$tip.label, common_sp))

traits_pruned <- traits %>%
  filter(Species1 %in% common_sp)

traits_pruned <- traits_pruned %>%
  mutate(mass_cr     = Mass_gram^(1/3),
         culmen_raw  = Beak_Length_Culmen_mm,
         width_raw   = Beak_Width_mm,
         depth_raw   = Beak_Depth_mm,
         Species     = Species1,
         culmen_norm = culmen_raw / mass_cr,
         width_norm  = width_raw  / mass_cr,
         depth_norm  = depth_raw  / mass_cr)

heat_df <- traits_pruned %>%
  select(Species, Clade,
         culmen_raw, width_raw, depth_raw,
         culmen_norm, width_norm, depth_norm) #%>%


groupInfo <- split(heat_df$Species, heat_df$Clade)
phy <- groupOTU(tree_pruned, groupInfo)


p <- ggtree(phy, aes(color=group), size = 0.01,layout="fan", open.angle = 180) #+ theme_tree2()

custom_colors1 <- c("Apterygiformes"  ="#2F0F0C", "Casuariiformes"   ="#3A130F", "Rheiformes"       ="#461713", "Struthioniformes"="#5D2E29", "Tinamiformes"       ="#73453F",
                    "Anseriformes"    ="#801A1A", "Galliformes"      ="#B33F3F", "Strisores"        ="#974364", "Columbiformes"   ="#BFBFFF", "Mesitornithiformes" ="#F0766D",
                    "Pterocliformes"  ="#B22222", "Gruiformes"       ="#556B2F", "Cuculiformes"     ="#808000", "Otidiformes"     ="#9ACD32", "Musophagiformes"    ="#8FBC8F",
                    "Charadriiformes" ="#008080", "Gaviiformes"      ="#72C0FF", "Pelecaniformes"   ="#ADD8E6", "Phaethontiformes"="#00BFCD", "Phoenicopteriformes"="#B4E0D0",
                    "Podicipediformes"="#5F9EA0", "Procellariiformes"="#4682B4", "Sphenisciformes"  ="#40416B", "Accipitriformes" ="#B8860B", "Bucerotiformes"     ="#A0522D",
                    "Coliiformes"     ="#DEB887", "Coraciiformes"    ="#FF8C00", "Leptosomatiformes"="#A0522D", "Piciformes"      ="#F4A460", "Strigiformes"       ="#2F4F4F",
                    "Trogoniformes"   ="#CD5C5C", "Cariamiformes"    ="#C23C42", "Falconiformes"    ="#E84B24", "Eurypygiformes"  ="#778899", "Opisthocomiformes"  ="#696969",
                    "Psittaciformes"  ="#FFD700", "Acanthisitti"     ="#228B22", "Tyranni"          ="#36A032", "Passeri"         ="#6bbf9f")
p2 <- p + 
  geom_fruit(geom=geom_bar, data=heat_df, mapping=aes(x=culmen_raw,y=Species,fill=Clade),
             orientation="y", offset = 0.05, pwidth = 0.1, stat="identity",) +
  geom_fruit(geom=geom_bar, data=heat_df, mapping=aes(x=width_raw,y=Species,fill=Clade),
             orientation="y", offset = 0.05, pwidth = 0.1, stat="identity") +
  geom_fruit(geom=geom_bar, data=heat_df, mapping=aes(x=depth_raw,y=Species,fill=Clade),
             orientation="y", offset = 0.05, pwidth = 0.1, stat="identity")+
  scale_fill_manual(values = custom_colors1)+scale_color_manual(values = custom_colors1)

p1 <- ggtree(phy, aes(color=group), size = 0.01,layout="fan", open.angle = 180) #+ theme_tree2()
p3 <- p1 + 
  geom_fruit(geom=geom_bar, data=heat_df, mapping=aes(x=culmen_norm,y=Species,fill=Clade),
             orientation="y", offset = 0.05, pwidth = 0.1, stat="identity",) +
  geom_fruit(geom=geom_bar, data=heat_df, mapping=aes(x=width_norm,y=Species,fill=Clade),
             orientation="y", offset = 0.05, pwidth = 0.1, stat="identity") +
  geom_fruit(geom=geom_bar, data=heat_df, mapping=aes(x=depth_norm,y=Species,fill=Clade),
             orientation="y", offset = 0.05, pwidth = 0.1, stat="identity")+
  scale_fill_manual(values = custom_colors1)+scale_color_manual(values = custom_colors1)
```
